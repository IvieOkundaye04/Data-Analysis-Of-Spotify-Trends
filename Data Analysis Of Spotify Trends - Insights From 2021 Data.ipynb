```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_tracks = pd.read_csv("tracks.csv")
```

```python
df_tracks.head()
```

```python
#check for null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
#to arrange the tracks in ascending levels of popularity
sorted_df = df_tracks.sort_values("popularity", ascending=True).head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
#to check for the most popular tracks 
most_popular = df_tracks.query("popularity>90", inplace = False).sort_values("popularity", ascending=False)
most_popular[:10]
```

```python
#to change the index to the release date of the tracks
df_tracks.set_index("release_date", inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
#the artist at the location of 18
df_tracks[["artists"]].iloc[18]
```

```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
corr_df = df_tracks.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson", numeric_only=True)
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="viridis", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y = "loudness", x = "energy", color="c").set(title="Correlation Between Loudness and Energy")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y = "energy", x = "acousticness", color="teal").set(title="Correlation Between Energy and Acousticness")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y = "popularity", x = "acousticness", color="purple").set(title="Correlation Between Popularity and Acousticness")
```

```python
df_tracks["dates"] = df_tracks.index.get_level_values("release_date")
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist", color="purple").set(title="Number Of Songs Per Year")
```

```python
total_dr = df_tracks.duration
fig_dims = (15,7)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False).set(title="Year vs. Duration") 
plt.xticks(rotation=90)
```

```python
total_dr = df_tracks.duration
sns.set_style(style="whitegrid")
gig_dims = (10, 5)
fig, ax = plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Year vs. Duration")
plt.xticks(rotation=60)
```

```python
df_genre = pd.read_csv("SpotifyFeatures.csv")
```

```python
df_genre.head(5)
```

```python
plt.title("Duration Of The Songs in Different Genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y="genre", x="duration_ms", data=df_genre)
plt.xlabel("Duration In Milliseconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,6))
famous = df_genre.sort_values("popularity", ascending=False).head(10)
sns.barplot(y="genre", x="popularity", data =famous).set(title="Top 5 Genres By Popularity")
```
